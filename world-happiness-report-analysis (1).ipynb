# Statistical tests

We are going with t test but not Z test because--If we had every possible countrys happiness score (not just those reported in the World Happiness Report), then we could consider that the true population standard deviation.


Even though we have all reported happiness scores, the World Happiness Report does not cover every country in the world. The missing countries create uncertainty, meaning your dataset is still a sample, not the entire population.

So, taking the standard deviation from the dataset and calling it the population standard deviation is incorrect, because:

The dataset doesnt include all possible countries.

Instead, the standard deviation you calculate from your dataset is an estimate of the true population standard deviation, which is why we use a T-test (not a Z-test) in such cases.

```python
sample_A=np.array(df_25_f['Ladder score'])
sample_A.shape
```

```python
sample_B=np.array(df_22_f['Ladder score'])
sample_B.shape
```

```python
sns.displot(df_22_f['Ladder score'], kde=True)
plt.show()
print(df_22_f['Ladder score'].skew())
```

```python
sns.displot(df_25_f['Ladder score'], kde=True)
plt.show()
print(df_25_f['Ladder score'].skew())
```

We have seen the distributions of both 2025 and 2022 happiness scores. It is not significantly different from normal distribution, so we can apply T-test to check the validity of null hypothesis.

Let us apply a T-test with null hypothesis as there is no significant difference between the happiness scores between 2025 and 2022. Alternate hypothesis is that there is significant difference between the happiness scores.

```python
# Let us see if there is significant difference between 2022 happiness and 2024 happiness scores 
# Is there a statistically significant difference in the mean happiness scores between 2022 and 2024?

from scipy import stats
import numpy as np


t_statistic, p_value = stats.ttest_ind(sample_A, sample_B)

alpha = 0.05
df = len(sample_A)+len(sample_B)-2

critical_t = stats.t.ppf(1 - alpha/2, df)

print("T-value:", t_statistic)
print("P-Value:", p_value)
print("Critical t-value:", critical_t)

print('With T-value')
if np.abs(t_statistic) >critical_t:
    print('There is significant difference between the two groups')
else:
    print('we fail to say if there is no significant difference between the two groups')

print('With P-value')
if p_value >alpha:
    print('No evidence to reject the null hypothesis that there is no significant difference between the two groups')
else:
    print('Evidence found to reject the null hypothesis that there is no significant difference between the two groups')
```

```python
# Let us see if there is significant difference between Europe and Asian countries Happiness scores

from scipy import stats
import numpy as np


t_statistic, p_value = stats.ttest_ind(asia_scores, europe_scores)

alpha = 0.05
df = len(asia_scores)+len(europe_scores)-2

critical_t = stats.t.ppf(1 - alpha/2, df)

print("T-value:", t_statistic)
print("P-Value:", p_value)
print("Critical t-value:", critical_t)

print('With T-value')
if np.abs(t_statistic) >critical_t:
    print('There is significant difference between Europe happpiness scores and Asia\'s happiness scores')
else:
    print('No significant difference found between two groups')

print('With P-value')
if p_value >alpha:
    print('No evidence to reject the null hypothesis that there is no significant difference between the two groups')
else:
    print('Evidence found to reject the null hypothesis that there is no significant difference between the two groups')
```
